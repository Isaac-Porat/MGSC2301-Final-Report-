Develop a Scatter Diagram showing the relationship between X and Y
- What does the scatter diagram indicate about a relationship between the X and Y?

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../student-mat.csv', delimiter=';')
```

```python
row_names = df.index.tolist()
column_names = df.columns.tolist()
```

```python
age_15_data = []
age_15_studytime_data = []
age_15_g1_data = []
age_15_g2_data = []
age_15_g3_data = []

for row in row_names:
    if df.loc[row, 'age'] == 15:
        studytime = df.loc[row, 'studytime']
        age_15_studytime_data.append(studytime)
        g1 = df.loc[row, 'G1']
        g2 = df.loc[row, 'G2']
        g3 = df.loc[row, 'G3']
        age_15_g1_data.append(g1)
        age_15_g2_data.append(g2)
        age_15_g3_data.append(g3)
        data_dict = {'G1': g1, 'G2': g2, 'G3': g3, 'Studytime': studytime}
        age_15_data.append(data_dict)
```

```python
X = age_15_studytime_data

max_g3 = max(age_15_g3_data)
age_15_g3_data_percentage = [(g3 / max_g3) * 100 for g3 in age_15_g3_data]
Y = age_15_g3_data_percentage

plt.scatter(X, Y)
plt.title('Student study time in relation to grade output')
plt.xlabel('Hours studied 1: < 2 hours, 2: 2 to 5 hours, 3: 5 to 10 hours, 4: > 10 hours')
plt.ylabel('Grade output out of 20')
plt.show()
```

Based on the scatter plot, we can see students who studied for more than 10 hours are in a tighter knit group between the 50% - 90% range. 

However, scatter diagrams do not best display the data due to the categorical nature of the independent variables.
